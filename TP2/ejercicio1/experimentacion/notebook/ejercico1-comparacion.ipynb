```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leyendo resultados
Primero, leemos el archivo de resultados de ejecución de bt. Para esto, vamos a abrirlo con Pandas y guardarlo en un DataFrame (df).

```python
df1 = pd.read_csv("tp2l/complejidades_array.csv")
df2 = pd.read_csv("tp2l/complejidades_tree.csv")
df3 = pd.read_csv("tp2l/complejidades_treePR.csv")
df4 = pd.read_csv("tp2l/complejidades_array_fijo.csv")
df5 = pd.read_csv("tp2l/complejidades_tree_fijo.csv")
df6 = pd.read_csv("tp2l/complejidades_treePR_fijo.csv")
df1['n'] = (df1['fila'] * df1["columna"])
df2['n'] = (df2['fila'] * df2["columna"])
df3['n'] = (df3['fila'] * df3["columna"])
df4['n'] = (df4['fila'] * df4["columna"])
df5['n'] = (df5['fila'] * df5["columna"])
df6['n'] = (df6['fila'] * df6["columna"])

```

```python
ax1 = sns.lineplot(x="n", y="tiempo", data=df1);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1 = sns.lineplot(x="n", y="tiempo", data=df3);
ax1.legend(["Array", "Tree", "Tree con Path Relinking"]);
plt.ylabel("tiempo (ms)");
plt.xlabel("h*w");
```

```python
#ax1 = sns.lineplot(x="n", y="tiempo", data=df4);
ax1 = sns.lineplot(x="n", y="tiempo", data=df5);
ax1 = sns.lineplot(x="n", y="tiempo", data=df6);
ax1.legend(["Tree", "Tree con Path Relinking"]);
plt.ylabel("tiempo (ms)");
plt.xlabel("h*w");
```

```python
#ax1 = sns.lineplot(x="n", y="tiempo", data=df4);
ax1 = sns.lineplot(x="n", y="tiempo", data=df1);
ax1 = sns.lineplot(x="n", y="tiempo", data=df4);
ax1.legend(["Array exp 1", "Array exp 2"]);
plt.ylabel("tiempo (ms)");
plt.xlabel("h*w");
```

```python
#ax1 = sns.lineplot(x="n", y="tiempo", data=df4);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1 = sns.lineplot(x="n", y="tiempo", data=df5);
ax1.legend(["Tree exp 1", "Tree exp 2"]);
plt.ylabel("tiempo (ms)");
plt.xlabel("h*w");
```

```python
#ax1 = sns.lineplot(x="n", y="tiempo", data=df4);
ax1 = sns.lineplot(x="n", y="tiempo", data=df3);
ax1 = sns.lineplot(x="n", y="tiempo", data=df6);
ax1.legend(["Tree con Path Relinking exp 1", "Tree con Path Relinking exp 2"]);
plt.ylabel("tiempo (ms)");
plt.xlabel("h*w");
```

```python
heat_df = df.pivot("fila", "columna", "t");
ax = sns.heatmap(heat_df)
ax.invert_yaxis()
```

```python
df["cota"] =((df["fila"])**2 * (df["columna"])**2);
# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df);
plt.xlabel("T(n)");
plt.ylabel("cota((h.w)^2)");
```
