# Experimentación de dos variables

## Importando bibliotecas
- <b>Pandas</b>: Nos sirve para leer datos y filtrarlos.
- <b>Seaborn</b>: Es una biblioteca para graficar que nos permite interactuar con datos de Pandas.

- <i>Numpy y Matplotlib son dependencias de Pandas.</i>


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leyendo resultados
Primero, leemos el archivo de resultados de ejecución de insertion sort. Para esto, vamos a abrirlo con Pandas y guardarlo en un DataFrame (df).

```python
df = pd.read_csv("resultado_merge.csv")
print(df)
```

## Gráfico en 3D
<i>No es buena idea en general porque no permite visualizar lo que se quiere transmitir de manera sencilla.</i>

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['n'], df['m'], df['t'], s=20)
ax.set_xlabel("n")
ax.set_ylabel("m")
ax.set_zlabel("tiempo (ms)")
plt.show()
```

## Heatmap
En este gráfico podemos ver el tiempo en función de las variables m y n en el color de los casilleros.
- Pro: Muestra todas las combinaciones de dos variables en un mismo gráfico.
- Pro: Da una idea de cómo es la relación de los tiempos en función de las variables.
- Con: No es tan fácil ver la función de tiempo (cuadrático? lineal?).
- Con: Solo un algoritmo en el gráfico.

```python
heat_df = df.pivot("n", "m", "t");
ax = sns.heatmap(heat_df)
ax.invert_yaxis()
```

## Gráfico de lineas con más de una serie
Cuando queremos mostrar cómo se comporta una función con dos variables se puede graficarla en función de una variable varias veces tomando ciertos valores representativos de la segunda.

A continuación graficamos el tiempo del algoritmo Merge en función de n. Esto lo hacemos para algunos valores de m.
Vemos que a medida que n crece, si fijamos m la función parece crecer linealmente. A su vez, mientras crece m pareciera que la pendiente de la función crece linealmente también.

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="t", data=df.loc[df['m'] == 10000]);
ax1 = sns.lineplot(x="n", y="t", data=df.loc[df['m'] == 40000]);
ax1 = sns.lineplot(x="n", y="t", data=df.loc[df['m'] == 70000]);
ax1 = sns.lineplot(x="n", y="t", data=df.loc[df['m'] == 100000]);
ax1.legend(["m=10000", "m=40000", "m=70000", "m=100000"]);
plt.ylabel("tiempo (ms)");
```

Ahora graficamos el tiempo del algoritmo Merge en función de m. Para algunos valores de n.
Vemos que sucede algo análogo al gráfico anterior.

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="m", y="t", data=df.loc[df['n'] == 10000]);
ax1 = sns.lineplot(x="m", y="t", data=df.loc[df['n'] == 40000]);
ax1 = sns.lineplot(x="m", y="t", data=df.loc[df['n'] == 70000]);
ax1 = sns.lineplot(x="m", y="t", data=df.loc[df['n'] == 100000]);
ax1.legend(["n=10000", "n=40000", "n=70000", "n=100000"]);
plt.ylabel("tiempo (ms)");
```

## Correlación entre funciones de más de una variable
Queremos ver de manera más precisa si hay una correlación entre la función de tiempo empírico y la cota. Para utilizar el coeficiente de correlación de Pearson tenemos que tener dos variables para comparar.<br />
Como nuestra cota teórica es O(n+m), vamos a tomar cada combinación de parámetros (n,m) y hacer un punto (n+m, T(n,m)). Luego, vamos a ver si existe una correlación entre ambas coordenadas.

```python
df["cota"] = df["n"] + df["m"];

# Grafico correlación.
ax1 = sns.lmplot(x='t', y='cota', data=df);
plt.xlabel("T(n,m)");
plt.ylabel("cota(n,m)");
```

Por último veamos el coeficiente de correlación de Pearson entre ambas series de datos.

```python
r = np.corrcoef(df["t"], df["cota"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

## ¿Cómo hubiesen sido los gráficos si nuestra implementación era O(nm)?

La idea ahora es ver que hubiese pasado si nuestra implementación daba un tiempo que no se ajustaba a la cota teórica. En este caso, vamos a ver que hay ciertos gráficos donde esto se puede ver y otros donde pasa desapercibido.

```python
df2 = pd.read_csv("resultado_merge_cuadratico.csv")
print(df2)
```

## Gráfico en 3D

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df2['n'], df2['m'], df2['t'], s=20)
ax.set_xlabel("n")
ax.set_ylabel("m")
ax.set_zlabel("tiempo (ms)")
plt.show()
```

En este gráfico no se puede percibir que la funcíon de tiempos es O(nm) en lugar de O(n+m), aunque se ve que el crecimiento aumenta a medida que n y m aumentan.

## Heatmap
Veamos si en el Heatmap hay señales de que la función de tiempos es O(nm).

```python
heat_df = df2.pivot("n", "m", "t");
ax = sns.heatmap(heat_df)
ax.invert_yaxis()
```

Tampoco parece haber una clara manera de distinguir si la función es O(nm) o O(n+m). Veamos que pasa con los gráficos de líneas.

## Gráficos de lineas con más de una serie

```python
fig = plt.figure()

# Graficamos el tiempo en función de n, con series variando m.
ax1 = fig.add_subplot(211)
ax1 = sns.lineplot(x="n", y="t", data=df2.loc[df2['m'] == 200]);
ax1 = sns.lineplot(x="n", y="t", data=df2.loc[df2['m'] == 800]);
ax1 = sns.lineplot(x="n", y="t", data=df2.loc[df2['m'] == 1600]);
ax1 = sns.lineplot(x="n", y="t", data=df2.loc[df2['m'] == 2000]);
ax1.legend(["m=200", "m=800", "m=1600", "m=2000"]);
plt.ylabel("tiempo (ms)");

# Graficamos el tiempo en función de m, con series variando n.
ax2 = fig.add_subplot(212)
ax2 = sns.lineplot(x="m", y="t", data=df2.loc[df2['n'] == 200]);
ax2 = sns.lineplot(x="m", y="t", data=df2.loc[df2['n'] == 800]);
ax2 = sns.lineplot(x="m", y="t", data=df2.loc[df2['n'] == 1600]);
ax2 = sns.lineplot(x="m", y="t", data=df2.loc[df2['n'] == 2000]);
ax2.legend(["n=200", "n=800", "n=1600", "n=2000"]);
plt.ylabel("tiempo (ms)");
```

En estos gráficos se empieza a notar que la función de tiempo empírico no parece tener la forma n+m. Esto se puede ver porque al fijar distintos valores de m o n cada vez más grandes, la función cada vez tiene pendiente más grande. Si la función fuese n+m, entoces al dejar fija una variable, la pendiente siempre sería la misma pero cambiaría la ordenada al origen como sucede más arriba en los anteriores experimentos.

## Correlación.

```python
df2["cota"] = df2["n"]+df2["m"];

# Grafico correlación.
ax1 = sns.lmplot(x='t', y='cota', data=df2);
plt.xlabel("T(n,m)");
plt.ylabel("cota(n,m)");

r = np.corrcoef(df2["t"], df2["cota"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

En este caso podemos ver que sigue existiendo una correlación (porque al aumentar n+m, también aumenta nm) pero en este caso es bastante más débil. Esto sugiere que n+m no parece ser una cota ajustada del problema.
