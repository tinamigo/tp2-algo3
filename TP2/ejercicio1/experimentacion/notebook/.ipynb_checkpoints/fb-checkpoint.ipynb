# Experimentación de dos variables

## Importando bibliotecas
- <b>Pandas</b>: Nos sirve para leer datos y filtrarlos.
- <b>Seaborn</b>: Es una biblioteca para graficar que nos permite interactuar con datos de Pandas.

- <i>Numpy y Matplotlib son dependencias de Pandas.</i>


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leyendo resultados
Primero, leemos el archivo de resultados de ejecución de insertion sort. Para esto, vamos a abrirlo con Pandas y guardarlo en un DataFrame (df).

```python
df = pd.read_csv("exp_todo_w_fb.csv")
print(df)
```

En el caso de fuerza bruta, deseamos corroborar que la cota teórica propuesta $O(n.2^n)$ resulta correcta. Por otra parte también nos parece pertinente analizar cuál es su dependencia con $W$ ya que el algoritmo realiza pequeñas podas a la hora de evaluar cada subconjunto (en el momento en el que realizo tanto la suma de tamaños como la de beneficios apenas obtengo una suma de capacidades mayor a $W$ dejo de considerar ese subconjunto y paso al siguiente). Nuestra hipótesis es que el valor de $W$ no va a afectar  el tiempo de ejecución para la experimentación en la que los valores de $w_{i}$ se encuentran entre $1$ y $W$, ya que al aumentar el W se generan instancias nuevas, en donde la distribución de los $w_{i}$ se mantiene debido a la utilización de valores aleatorios que provienen de una distribución uniforme entre $1$ y $W$ y por lo tanto la magnitud de casos que son descartados es similar.

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

Podemos ver que a medida que aumenta $n$ aumenta el tiempo de ejecución, como también podemos observar que el crecimiento del tamaño de $W$ no necesariamente determina un crecimiento en el tiempo de ejecución. Esto concuerda con nuestra hipótesis inicial ya que muestra la independecia de $W$ con respecto al tiempo de ejecución en las condiciones dadas por el experimento. Para observar esto de forma más precisa graficaremos distintos $n$ en un gráfico de tiempo de ejecución en función de $W$. En cuanto a la cota, parecería ser acertado decir que para cada $W$, el tiempo en función de $n$ se asemeja a una función exponencial, sin embargo faltaría realizar un análisis más detallado para comprobarlo. Esto lo realizaremos más adelante.

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 22]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 28]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 31]);
ax1.legend(["n=10", "n=22", "n=28" , "n=31"]);
plt.ylabel("tiempo (ms)");
```

El gráfico muestra que para distintos valores de $n$, los tiempos de ejecución para distintos $W$ no muestran un comportamiento determinado, si no más bien parece ser errático, tomando valores menores o mayores sin depender del valor específico de $W$. Esto es atribuible al comportamiento distinto que se genera debido a diversas instancias, lo cual corrobora nuestra hipótesis original. Vale destacar que seguimos notando que para mayores $n$ vemos un crecimiento en los valores que toman los tiempos de ejecución, si bien de esta forma es poco notable si se trata de una función exponencial o no.


```python
df_cuadratico = pd.DataFrame({"n": [i for i in range(1, 31)], "IS": [0.4e-5*(2**(i) * (i)) for i in range(1, 31)]});
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df);
ax1 = sns.lineplot(x="n", y="IS", data=df_cuadratico);
ax1.legend(["Experimentación 1", "0.4e-5.(n.2^(n))"]);
plt.ylabel("tiempo (ms)");
```

Notamos en el gráfico que $T(n)$ está acotada por la cota teórica y la misma parece estar ajustada. Veamos mediante el gráfico de correlación entre la cota teórica y el tiempo de ejecución $T(n)$ que se trata de hecho una cota apropiada.

```python
df["cota"] = df["n"] * (2**(df["n"]));

# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df.loc[df["W"] == 20]);
ax1 = sns.lmplot(x='tiempo', y='cota', data=df.loc[df["W"] == 70]);
plt.xlabel("T(n)");
plt.ylabel("cota(n.2^n)");
```

Elegimos representar el gráfico de correlación para dos sets distintos de data, correspondientes a $W=20$ y $W=70$, ya que estos representan dos extremos opuestos en el rango de $W$ utilizado. En ambos gráficos los puntos parecen estar alineados, por lo que podríamos decir que existe una fuerte correlación entre $T(n)$ y la cota teórica. Esto podemos corroborarlo con el coeficiente de Pearson calculado para todos los $W$, cuyo valor es 0.9873362544932353 (aproximadamente 1).

```python
df_coef_1 = df
r = np.corrcoef(df_coef_1["tiempo"], df_coef_1["cota"])[0, 1];
print("El coeficiente de correlación de Pearson para la primera experimentación es: r="+str(r))
```

Luego de realizar el análisis de la cota superior nos pareció interesante realizar el mismo análisis de la posible independencia de $W$ en el marco de la segunda experimentación. Bajo estas condiciones, si nuestra hipótesis es correcta, suponemos que al tratarse de una misma instancia de $n$ elementos con un rango fijo para los $w_{i}$ para cada $W$, cuando $W$ aumente habrá menos casos en los que la suma de tamaños de cada subconjunto sea en algún punto mayor que $W$ y por lo tanto habrá menos podas. Es decir, se ejecutarán más operaciones en promedio por subconjunto, viéndose así afectada la eficiencia del algoritmo.

```python
df2 = pd.read_csv("exp_todo_w_benefit_size_fijos_fb.csv")
print(df2)
```

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df['n'] == 22]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df['n'] == 28]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df['n'] == 31]);
ax1.legend(["n=10", "n=22", "n=28" , "n=31"]);
plt.ylabel("tiempo (ms)");
```

Vemos en el primer gráfico que en este caso el valor de $W$ afecta el tiempo de ejecución del algoritmo, ya que a mayor $W$ notamos un mayor $T(n)$. En cuanto al segundo gráfico, apreciamos que se nota un claro crecimiento del tiempo de ejecución para una misma instancia en la que se varía el $W$. Sin embargo, la curvatura convexa de la función para los distintos $n$ pareciera indicar que existe un estancamiento en el crecimiento de la misma, lo que asociamos al valor de $W$ a partir del cual la suma de los $w_{i}$ es menor que $W$ para todos los subconjuntos posibles (es decir, no existen podas por lo que ya no puede empeorar el algoritmo).

Comparemos los tiempos de ejecución de ambos experimentos:

```python
#df_cuadratico = pd.DataFrame({"n": [i for i in range(1, 31)], "IS": [0.4e-5*(2**(i) * (i)) for i in range(1, 31)]});
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
#ax1 = sns.lineplot(x="n", y="IS", data=df_cuadratico);
ax1.legend(["Experimentación 1", "Experimentación 2"]);
plt.ylabel("tiempo (ms)");
```

```python
df_cuadratico_1 = pd.DataFrame({"n": [i for i in range(1, 31)], "IS": [0.1e-4*(2**(i) * (i)) for i in range(1, 31)]});
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1 = sns.lineplot(x="n", y="IS", data=df_cuadratico_1);
ax1.legend(["Experimentación 2", "0.1e-4.(n.2^(n))"]);
plt.ylabel("tiempo (ms)");
```

La comparación entre ambos experimentos muestra lo que esperábamos, ya que la segunda experimentación se encuentra por encima de la primera, debiéndose esto a la menor cantidad de podas que realiza. Sin embargo, en el segundo gráfico vemos que de todas formas se sigue tratando de un crecimiento exponencial, resultando apropiada la cota teórica de $O(2^n * n)$

```python
df["cota"] = df["n"] * (2**(df["n"]));
df2["cota"] = df2["n"] * (2**(df2["n"]));

# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df.loc[df["W"]==100]);
ax1 = sns.lmplot(x='tiempo', y='cota', data=df2.loc[df2["W"]==100]);
plt.xlabel("T(n)");
plt.ylabel("cota(n.2^n)");
```

Por último veamos el coeficiente de correlación de Pearson entre ambas series de datos.

```python
df_coef_1 = df.loc[df["W"]==100]
df_coef_2 = df2.loc[df2["W"]==100]
r = np.corrcoef(df_coef_1["tiempo"], df_coef_1["cota"])[0, 1];
print("El coeficiente de correlación de Pearson para la primera experimentación es: r="+str(r))
r2 = np.corrcoef(df_coef_2["tiempo"], df_coef_2["cota"])[0, 1];
print("El coeficiente de correlación de Pearson para la segunda experimentación es: r="+str(r2))
```

Corroboramos mediante los gráficos de correlación utilizando un valor fijo de $W = 70$ ya que es en este valor en donde en la segunda experimentación más notamos la diferencia en cantidad de podas realizadas, acercándonos más al peor caso. El coeficiente del segundo es mayor que el primero, confirmando la relación sospechada (0.9996759151761935 > 0.9999074969270538).
