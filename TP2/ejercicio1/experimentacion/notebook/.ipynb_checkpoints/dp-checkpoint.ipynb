# Experimentación de dos variables

## Importando bibliotecas
- <b>Pandas</b>: Nos sirve para leer datos y filtrarlos.
- <b>Seaborn</b>: Es una biblioteca para graficar que nos permite interactuar con datos de Pandas.

- <i>Numpy y Matplotlib son dependencias de Pandas.</i>


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leyendo resultados
Primero, leemos el archivo de resultados de ejecución de bt. Para esto, vamos a abrirlo con Pandas y guardarlo en un DataFrame (df).

```python
df = pd.read_csv("dp_variante_200.csv")
print(df)
```

Para el primer experimento con el algoritmo de programación dinámica, nuestra hipótesis es que la cota teórica propuesta de $O(n * W)$ va a resultar apropiada ya que realizamos $n$ iteraciones y en cada una de ellas se paga el costo de copiar el vector de tamaño $W$ lo que nos garantiza una complejidad de $n * W$. Lo que puede afectar el tiempo de ejecución es el hecho de que no recorremos la totalidad del vector sino que guardamos los índices que se encuentran de hecho definidos en él. En cuanto a ambas variables, esperamos ver que el tiempo de ejecución dependa de ambas ya que su cota teórica se define a partir de tanto $n$ como $W$. 

En cuanto a la comparación entre ambas experimentaciones, esperamos que en la primera, al tener un rango de $w_{i}$ variantes, existan más valores que al tratar de definirlos en el vector correspondiente del siguiente elemento se pasen de $W$ y por lo tanto esto resulte en una menor cantidad de valores definidos en nuestro diccionario. De hecho, si tenemos en cuenta que la distribución de los $w_{i}$ es uniforme entre $1$ y $W$ entonces la esperanza de $w_{i}$ es igual a $W/2$ por lo que la esperanza de la suma de tres elementos cualesquiera ya es mayor a $W$ por lo que se irán de rango. Por ende, el tiempo de ejecución del experimento 1 será menor que el del segundo, ya que este al tener valores fijos de $w_{i}$ para $W$ mayores tendrá menos valores que se pasan de $W$ y por lo tanto definiré más valores en la matriz (o vector en este caso).

Vale notar que debido a que nuestra cota es lineal con respecto a $n$, utilizamos valores mayores de $n$ para poder percibir mejor el comportamiento del algoritmo a medida que $n$ toma valores mucho mayores. Además, como este algoritmo tiene una cota teórica mucho mejor que los anteriores, pudimos permitirnos utilizar $n$ mayores y correr los experimentos en un tiempo razonable (cosa que no podíamos hacer con fuerza bruta).

Pasemos a analizar los datos de la primera experimentación. 

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 61]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 103]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 199]);
ax1.legend(["n=10", "n=61", "n=103" , "n=199"]);
plt.ylabel("tiempo (ms)");
```

En el primer gráfico vemos que fijando distintos valores de $W$ al incrementar $n$ crece el tiempo de ejecución, notando entonces la dependencia de $T(n)$ de $n$, pareciendo una relación lineal si bien notamos altibajos. Análogamente en el segundo gráfico donde fijamos $n$ se ve la relación del tiempo de ejecución con $W$ y seguimos notando su linealidad como en el anterior caso. Las dicidencias con el modelo lineal que notamos se pueden deber a la generación de instancias aleatorias para cada $n$ y $W$ distinto, variando entonces la cantidad de valores definidos en la matriz.

Para poder ver si existe correlación con la cota teórica del algoritmo, confeccionaremos el gráfico de correlación entre $T(n)$ y la función $n * W$.

```python
df["cota"] = (df["n"]) * (df["W"]);
# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df);
plt.xlabel("T(n)");
plt.ylabel("cota(n.W)");
```

```python
r = np.corrcoef(df["tiempo"], df["cota"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

En el gráfico de correlación podemos notar que existe una gran conglomeración de puntos cercanos a una alineación. Por esta razón y por su coeficiente de Pearson (0,97..) podemos concluir que la cota es apropiada. Esto se condice con lo hipotetizado.

A continuación, analizaremos los resultados obtenidos para la segunda experimentación. Realizaremos los mismos gráficos que en el caso anterior para poder realizar una comparación entre ellos.

```python
df2 = pd.read_csv("dp_fijos_200.csv")
```

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 61]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 103]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 199]);
ax1.legend(["n=10", "n=61", "n=103" , "n=199"]);
plt.ylabel("tiempo (ms)");
```

En este experimento, seguimos notando la misma relación lineal tanto en el caso de $n$ como de $W$, lo que indica que sigue respetando la cota. Sin embargo, notamos ciertas diferencias. Por un lado, en el gráfico de la izquierda  observamos que para distintos $W$ los altibajos que se notan en la función parecen tener un mismo comportamiento marcado. Es decir, que los tiempos de ejecución se incrementan y decrementan para los mismos $n$. Esta diferencia la atribuimos a que en esta experimentación se utilizaron las mismas instancias para un valor determinado de $n$ y luego se varió $W$, por lo que tiene sentido que si el tiempo de ejecución aumenta o disminuye con respecto al anterior $n$ para un $W$ también lo haga para el resto de los $W$.
En el gráfico de la derecha vemos que son menos distinguibles las dicidencias con el modelo lineal, lo cual se condice con nuestra hipótesis inicial ya que en este caso se definen más valores en la matriz como explicamos previamente. 

Veamos el gráfico de correlación y el coeficiente de Pearson para corroborar lo planteado.

```python
df2["cota"] = (df2["n"]) * (df2["W"]);

# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df2);
plt.xlabel("T(n)");
plt.ylabel("cota(n.W)");
```

```python
r = np.corrcoef(df2["tiempo"], df2["cota"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

Vemos en el gráfico de correlación que la conglomeración de puntos se encuentran menos dispersos que en el caso anterior, lo cual indica una mejor correlación; esto lo corrobora el valor del coeficiente, de mayor valor también que en la experimentación 1. 

Por último, presentamos los datos en un gráfico que permite verificar la relación planteada en nuestra hipótesis entre el tiempo de ejecución de la primera y segunda experimentación. 

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1.legend(["Experimentación 1", "Experimentación 2"]);
plt.ylabel("tiempo (ms)");
```
