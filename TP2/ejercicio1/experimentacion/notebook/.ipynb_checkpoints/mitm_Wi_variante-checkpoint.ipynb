# Experimentación de dos variables

## Importando bibliotecas

- <b>Seaborn</b>: Es una biblioteca para graficar que nos permite interactuar con datos de Pandas.

- <i>Numpy y Matplotlib son dependencias de Pandas.</i>


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leyendo resultados
Primero, leemos el archivo de resultados de ejecución de insertion sort. Para esto, vamos a abrirlo con Pandas y guardarlo en un DataFrame (df).

```python
df = pd.read_csv("exp_todo_w_mitm.csv")
print(df)
```

A continuación graficamos el tiempo del algoritmo Merge en función de n. Esto lo hacemos para algunos valores de m.
Vemos que a medida que n crece, si fijamos m la función parece crecer linealmente. A su vez, mientras crece m pareciera que la pendiente de la función crece linealmente también.

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

En este gráfico podemos ver que a medida que $n$ crece también lo hace el tiempo de ejecución. La curvatura parecería indicar una relacion exponencial, lo cual se condice con la cota teórica propuesta$(O(n.2^{\frac{n}{2}})$. Por otra parte, podríamos decir que lo que tarda en ejecutarse es independiente del valor de $W$, ya que para distintos valores del mismo no parece haber mucha diferencia en $T(n)$. Esto lo notamos en la similitud de su crecimiento en el gráfico.
Para analizar mejor la independecia de $W$ y la cota de tiempo, vamos a analizar la relacion recíproca, es decir, para un determinado valor de $n$ variaremos $W$ y veremos su comportamiento.

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 22]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 28]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 31]);
ax1.legend(["n=10", "n=22", "n=28" , "n=31"]);
plt.ylabel("tiempo (ms)");
```

En este gráfico, podemos ver que, como mencionamos antes, no notamos una relacion directa entre el valor de $n$ y $W$ ya que en el gráfico para cada $n$ no vemos un comportamiento homogéneo a medida que crece $W$(no crece ni decrece proporcionalmente) si no más bien observamos un comportamiento errático. 
Asociamos este comportamiento errático a que para cada $W$ se generan instancias aleatorias independientes entre sí(donde para cada elemento su $w_{i}$ varía entre 1 y $W$) y es por esto que el tiempo de ejecución no es siempre similar, ya que depende mucho de qué instancia se trate. En particular, puede suceder que en ciertas instancias cuando se aplique fuerza bruta a cada mitad, su $output$ sea de tamaño menor a otras, debido a que sólo devuelve instancias que vale la pena representar (aquellas cuya suma de tamaño es menor que $W$). Por lo tanto, el tamaño de cada mitad que luego unificaré se puede ver muy reducido en comparación con $2^{\frac{n}{2}}$ cambiando el tiempo de ejecución de diferentes instancias.



Ahora sí, para poder analizar con mayor precisión si la cota teórica se condice con su tiempo de ejecución vamos a realizar para cada n un promedio de todas las ejecuciones para los distintos $W$ entre 20 y 100 con saltos de 5 en 5(20,25,...,100). Para poder tener una apreciació correcta representaremos en el mismo gráfico una función de orden $2^{\frac{n}{2}}. \frac{n}{2}$ multiplicado por una constante para que ambos se encuentren en una escala similar. 

```python
df_cuadratico = pd.DataFrame({"n": [i for i in range(1, 30)], "IS": [0.2e-4*(2**(i/2) * (i/2)) for i in range(1, 30)]});
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df);
ax1 = sns.lineplot(x="n", y="IS", data=df_cuadratico);
ax1.legend(["promedio de W", "0.2e-4.(n.2^(n/2))"]);
plt.ylabel("tiempo (ms)");
```

Como podemos ver representado, notamos que la función que relaciona el tiempo con el tamaño de la entrada($n$) pareciera tener un orden similar a la cota teórica propuesta ($O(\frac{n}{2}.2^{\frac{n}{2}})$).

## texto nuestro



Para poder establecer con seguridad que la cota propuesta es correcta analizaremos la correlación existente entre $T(n)$ y la cota teórica. 


```python
df["cota"] = df["n"] * (2**(df["n"]/2));

# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df.loc[df["W"]==20]);
plt.xlabel("T(n)");
plt.ylabel("cota(n.2^n)");
```

```python
df["cota"] = df["n"] * (2**(df["n"]/2));

# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df.loc[df["W"]==100]);
plt.xlabel("T(n)");
plt.ylabel("cota(n.2^n)");
```

Por último veamos el coeficiente de correlación de Pearson entre ambas series de datos.

```python
r = np.corrcoef(df["tiempo"], df["cota"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

Vemos en ambos gráficos de correlación (para un $W$ de valor 20 y 100 ya que se trata de casos extremos) que los puntos se encuentran alineados, apuntando a la existencia de una correlación entre ambas variables. Como el coeficiente de correlación de Pearson tiene un valor de 0.981... podemos decir que la cota estimada es correcta.

```python
df2 = pd.read_csv("exp_todo_w_benefit_size_fijos_mitm.csv")
print(df2)
```

Otra duda que nos surge en cuanto al comportamiento del tiempo de ejecución del algoritmo es de qué otros parámetros podría estar dependiendo. En el experimento anterior, los valores del tamaño de cada elemento de cada instancia variaban entre 1 y el $W$ que le correspondía. Planteamos que si fijamos los valores del tamaño en un rango fijo y menor a todos los W utilizados, esto podría afectar al tiempo final. Nuestra hipótesis es que, debido a que el algoritmo de fuerza bruta que se aplica a ambas mitades del conjunto original descarta aquellos subconjuntos cuya suma de capacidades exceda $W$, para valores de $W$ mayores que permitan una mayor cantidad de subconjuntos el tiempo de ejecución se verá afectado. Resumiendo, si fijamos los $w_{i}$ en un rango, entonces a mayor $W$, mayor tiempo de ejecución. 
Para esto, tomamos valores de $n$ entre 1 y 31 (de 3 en 3), con valores de $W$ entre 20 y 100 (de 5 en 5) y fijando $w_{i}$ en un rango entre 1 y 20 para cada $n$ fijo. A continuación, presentamos los resultados en dos gráficos distintos:

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

Como podemos apreciar en el gráfico, vemos que para $W$ mayores hay una mayor diferencia en tiempo de ejecución, que se distingue más en los valores de $n$ más grandes. Esto coincide con lo que esperábamos encontrar ya que con un mayor $W$ vemos que la curva se nota mucho más pronunciada. Para corrobar aún más, ya que para un mismo $n$ contábamos con la misma instancia que sólo varía su $W$, realizamos el gráfico con la relación inversa, es decir, vemos para ciertos valores representativos de $n$ el tiempo en función de $W$.

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 22]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 28]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 31]);
ax1.legend(["n=10", "n=22", "n=28" , "n=31"]);
plt.ylabel("tiempo (ms)");
```

En este gráfico podemos ver más claramente que para una misma instancia el valor de $W$ afecta proporcionalmente al tiempo de ejecución del algoritmo. Podríamos decir que notamos una relación directa entre ambas variables. Para constatarlo, realizaremos el gráfico de correlación entre $T(n)$ y $W$ para un $n$ fijo (elegimos el valor 31 ya que consideramos que es donde más se notaría la relación). 

```python
# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='W', data=df2.loc[df2['n']==31]);
plt.xlabel("T(n)");
plt.ylabel("W");
```

```python
df3 = df2.loc[df2['n']==31];
r = np.corrcoef(df3["tiempo"], df3["W"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

Podemos ver en el gráfico de correlación, entonces, que pareciera existir dicha relación, corroborado por el valor de coeficiente de correlación de Pearson. Como este es cercano a 1, aseguramos que el $W$ afecta al tiempo de ejecución causando que este mismo aumente. 

Por otra parte, notamos en los gráficos anteriores que el valor del tiempo de ejecución también son mucho mayores que aquellos que habíamos obtenido en la experimentación con los distintos $w_{i}$ variando entre 1 y $W$. Para poder observar mejor esta diferencia, graficamos a continuación ambos experimentos:

```python
ax1 = sns.lineplot(x="n", y="tiempo", data=df);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1.legend(["$w_{i}$ variante", "$w_{i}$ fijo entre 1 y 20"]);
plt.ylabel("tiempo (ms)");
```

Como podemos ver, existe una gran diferencia entre los resultados de ambos experimentos. Notamos que a medida que el $n$ crece los tiempos relacionados con cada instancia crecen mucho más rapidamente con los tamaños fijados en comparación con la opción de variarlos con el $W$ correspondiente. Creemos que esto se debe a nuestra hipótesis planteada; es decir, que con tamaños $w_{i}$ más pequeños existen más subconjuntos que logran no sobrepasar $W$ y por lo tanto fuerza bruta no descarta tantos casos como en la otra experimentación, causando que la combinación de ambas mitades sea más costosa.

Nótese que para este algoritmo no consideramos cambiar los rangos del beneficio de cada elemento; esto es así ya que como no se realiza ninguna poda por optimalidad (es decir, en ningún momento de mi algoritmo considero el beneficio para tomar una decisión) y por lo tanto los valores de beneficio no afectan al mismo.
