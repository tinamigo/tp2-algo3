```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("exp_todo_w_fb.csv")
df2 = pd.read_csv("exp_todo_w_mitm.csv")
df3 = pd.read_csv("exp_todo_w_bt.csv")
df4 = pd.read_csv("exp_todo_w_dp.csv")
df5 = pd.read_csv("exp_todo_w_benefit_size_fijos_fb.csv")
df6 = pd.read_csv("exp_todo_w_benefit_size_fijos_mitm.csv")
df7 = pd.read_csv("exp_todo_w_benefit_size_fijos_bt.csv")
df8 = pd.read_csv("exp_todo_w_benefit_size_fijos_dp.csv")
```

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df1);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1 = sns.lineplot(x="n", y="tiempo", data=df3);
ax1 = sns.lineplot(x="n", y="tiempo", data=df4, size = 1);
ax1.legend(["Fuerza bruta", "Meet-in-the-middle", "Backtracking", "Programación dinámica"]);
plt.ylabel("tiempo (ms)");
```
