```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leyendo resultados
Primero, leemos el archivo de resultados de ejecución de bt. Para esto, vamos a abrirlo con Pandas y guardarlo en un DataFrame (df).

```python
df = pd.read_csv("exp_todo_w_bt.csv")
print(df)
```

Para este experimento con el algoritmo de backtracking, nuestra hipótesis es que la cota teórica del peor caso no resultará ajustada para un conjunto de instancias con distribución aleatoria uniforme entre determinados valores ya que en el caso promedio no debe recorrer todos los nodos del árbol de backtracking (ya que se realizan podas). Por otra parte, en cuanto a cómo influye el valor de $W$, suponemos que como la eficiencia del algoritmo depende de las podas que se realicen, entonces su tiempo de ejecución dependerá también de $W$ ya que es un valor que afecta a las podas realizadas. 

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df.loc[df['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

Vemos en el gráfico que para distintos valores de $W$ no hay una relación que se mantenga a medida que varía $n$. Es decir que en esta experimentación parecería que $W$ no influye en el tiempo de ejecución, por lo que nuestra hipótesis inicial no era correcta. Otra hipótesis superadora acerca de este comportamiento es que al variar los distintos tamaños de $w_{i}$ de los elementos entre $1$ y $W$ para distintos tamaños de $W$ se generan instancias similares en cuanto al comportamiento de la poda de factibilidad ya que un mayor $W$ no implica que haya menos podas o que se realice la poda más tarde en el árbol de decisiones.
Para probar esta hipótesis, realizaremos el experimento variando los valores de $w_{i}$ entre $1$ y $20$ (utilizando la misma instancia con $n$ elementos para cada $W$) ya que de esta forma todos los valores de tamaños son menores que todos los $W$ utilizados en la experimentación y podremos ver de mejor manera como afecta el valor de $W$ a una misma instancia.

```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 22]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 28]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df.loc[df['n'] == 31]);
ax1.legend(["n=10", "n=22", "n=28" , "n=31"]);
plt.ylabel("tiempo (ms)");
```

Al igual que en $meet$-$in$-$the$-$middle$, y como mencionamos en el gráfico anterior, notamos un comportamiento errático para un mismo $n$ con distintos $W$. Esto refuerza nuestra hipótesis del párrafo anterior, por lo que utilizaremos la experimentación descrita para corroborarla o refutarla.

En cuanto al análisis de la cota, en los gráficos anteriores si bien notamos que para mayores $n$ se obtienen mayores valores de tiempo de ejecución, no observamos un crecimiento consistente, pareciendo entonces no tener un comportamiento netamente exponencial. De hecho, para mismos valores de $n$ con distintos $W$ notamos una gran diferencia en tiempos de ejecución. Esto lo volvemos a atribuir a que para diferentes instancias varían las podas realizadas tanto de factibilidad como de optimalidad repercutiendo considerablemente en la eficiencia del algoritmo. Recordemos que una de las razones por la que backtracking resulta más eficiente que fuerza bruta es que en vez de analizar todas las posibles soluciones, se realizan podas en el momento en el que se puede determinar que no se trata de una solución válida, reduciendo así la cantidad de soluciones analizada y siendo cuando esto no sucede el peor caso para el algoritmo de backtracking. Por lo tanto, una menor cantidad de podas implica un mayor tiempo de ejecución (el tiempo se acercará más a la complejidad teórica del peor caso).

Vemos que la cota $O(2^n)$ resulta estar muy por encima de nuestro promedio de tiempos de ejecución, parecería que en un caso promedio no resulta una cota ajustada, alejándose de lo que sería el peor caso. Tanto el gráfico de correlación como el coeficiente de Pearson parecen corrobar este hecho, ya que si bien su valor es positivo y sobrepasa el 0,5 no se acerca lo suficiente a 1 como para poder afirmar que existe una correlación tan fuerte. Esto lo relacionamos con las podas realizadas, que disminuyen los casos analizados, impactando positivamente en la performance de nuestro algoritmo. Por esta razón, vamos a intentar acercarnos a una cota que parezca más apropiada para los casos promedios de esta experimentación y presentaremos los gráficos correspondientes, incluyendo el de correlación y el coeficiente de Pearson para dicha cota. Sin embargo, para poder verificar que la cota es apropiada en el peor caso vamos a utilizar la experimentación ya mencionada anteriormente (variando $w_{i}$ en un rango fijo menor a $W$). Vale destacar que si bien $O(2^n)$ no parece ser una cota ajustada en el caso promedio, esto no quiere decir que no sea una cota válida para el peor caso (caso en el que no existen podas, que exploraremos en la experimentación mencionada). La nueva cota propuesta es $O(2^\frac{n}{4})$ $=$ $O(2^n)$ y hemos utilizado una constante para lograr escalar el gráfico.

```python
df_cuadratico = pd.DataFrame({"n": [i for i in range(1, 30)], "IS": [0.3e-2*(2**(i/4)) for i in range(1, 30)]});
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df);
ax1 = sns.lineplot(x="n", y="IS", data=df_cuadratico);
ax1.legend(["promedio de W", "0.3e-2.(2^(n/4))"]);
plt.ylabel("tiempo (ms)");
```

```python
df["cota"] = 2**((df["n"])/4);
df_corr = df.loc[df["W"]==20]
# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df_corr);
plt.xlabel("T(n)");
plt.ylabel("cota(n.2^(n/4))");
```

```python
df["cota"] = 2**((df["n"])/4);
df_corr2 = df.loc[df["W"]==100]
# Grafico correlación.
ax1 = sns.lmplot(x='tiempo', y='cota', data=df_corr2);
plt.xlabel("T(n)");
plt.ylabel("cota(n.2^(n/4))");
```

```python
r = np.corrcoef(df["tiempo"], df["cota"])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

Podemos ver que la nueva cota propuesta se ajusta más al caso promedio, observable a través de los gráficos con puntos ligeramente más alineados y el coeficiente más cercano a 1, lo que indica una correlación más fuerte que para la anterior cota. Sin embargo, no termina mostrando una correlación perfecta.

```python
df2 = pd.read_csv("exp_todo_w_benefit_size_fijos_bt.csv")
```

Pasemos a la siguiente experimentación, como ya mencionamos previamente en esta experimentación los $w_{i}$ serán asignados de forma aleatoria para valores entre $1$ y $20$, variando los valores de $W$ para una misma instancia de $n$ elementos. Nuestra hipótesis para esta nueva experimentación es que, a diferencia del caso anterior, veremos una relación más marcada entre el valor de $W$ y el tiempo de ejecución. Esto, suponemos, es así ya que para una misma instancia, el distinto valor de capacidad máxima influye de manera directa en la cantidad de subconjuntos cuya suma superará $W$ y, por lo tanto, serán podados en una determinada posición del árbol de decisiones. Es decir, si menos subconjuntos llegan a sumar $W$ entonces a medida que se forman las soluciones parciales de dichos subconjuntos estos no serán descartados, confiando entonces el algoritmo únicamente en la poda por optimalidad como opción para no considerar la solución hasta el final (siendo este el peor caso ya que me acerco a realizar las $2^n$ llamadas, resultando en una complejidad de O($2^n$)).

```python
# Graficamos el tiempo en función de n, con series variando m.
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 20]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 40]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 70]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2['W'] == 100]);
ax1.legend(["W=20", "W=40", "W=70", "W=100"]);
plt.ylabel("tiempo (ms)");
```

En el gráfico vemos que, a diferencia de la anterior experimentación, en esta ocasión sí notamos una relación marcada entre el valor de $W$ y cómo este influye en un mayor tiempo de ejecución para una misma cantidad $n$ de elementos. Esto apunta a que $W$ está afectando la performance del algoritmo, tal como planteamos en nuestra hipótesis. Para valores mayores de $W$ vemos cómo cada vez va creciendo más el tiempo de ejecución a medida que crece $n$, creciendo cada vez más la diferencia entre distintos valores de $W$. Si bien con un $n$ cercano a $30$ vemos que con $W$ menores como $20$ y $40$ la diferencia entre ellos parece ser mínima, ya con un valor de $70$ notamos un salto cercano a $40ms$ en tiempo de ejecución, mientras que con un valor de $100$ esta diferencia ya pasa a ser aproximadamente de $200ms$, el quíntuple del salto anterior (notemos que en estos dos últimos ejemplos $W$ está variando lo mismo, sin embargo su impacto en el tiempo de ejecución es muchísimo mayor). Esto indicaría incluso que la relación del impacto de $W$ no es lineal. 

En cuanto a los valores del tiempo a medida que crece $n$ seguimos notando que a medida que $n$ crece, también lo hace el tiempo de ejecución. Es notable también apreciar que a medida que $W$ crece vemos un comportamiento que parece ser más marcadamente exponencial, lo que podría estar indicando que a medida que $W$ es mayor, al ser peor la poda de factibilidad, nos acercamos más a un peor caso en el que se realizan las $2^n$ llamadas recursivas y por lo tanto su complejidad de $O(2^n)$ es más ajustada. En los casos en los que la poda de factibilidad es más efectiva (menor $W$), la cota es menos acertada, ya que se realizan muchas menos llamadas en el árbol de decisiones y, por lo tanto, la cota resulta muy por encima de los valores promedios reales. 

Veamos qué sucede cuando graficamos la relación inversa; para un mismo $n$, el tiempo de ejecución en función de $W$.


```python
# Graficamos el tiempo en función de m, con series variando n.
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 10]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 22]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 28]);
ax1 = sns.lineplot(x="W", y="tiempo", data=df2.loc[df2['n'] == 31]);
ax1.legend(["n=10", "n=22", "n=28" , "n=31"]);
plt.ylabel("tiempo (ms)");
```

En este gráfico vemos claramente que existe una relación entre el valor de $W$ y el tiempo de ejecución; a medida que el primero crece, también lo hace el segundo. Notemos nuevamente que parece tratarse de una relación no lineal, sino más bien de orden exponencial. Nos interesa corroborar, por lo tanto, la intuición que obtuvimos de ambos gráficos. Tanto la idea de que a medida que $W$ crece la ejecución se acerca más a la de un peor caso con una cota de $O(2^n)$ como la idea de que la relación entre el tiempo de ejecución y $W$ es exponencial para una misma instancia de $n$ elementos (A QUE PODRIAMOS ATRIBUIR ESTO? ). Para ello, miremos primeramente en un gráfico cómo es la relación de los distintos $W$ con la cota teórica de $2^n$.

```python
df_cuadratico2 = pd.DataFrame({"n": [i for i in range(1, 31)], "IS": [0.9e-6*(2**(i)) for i in range(1, 31)]});
# Graficamos el tiempo en función de n, con series variando m.
dfW20 = df2.loc[df2["W"]==20] 
dfW50 = df2.loc[df2["W"]==50]
dfW100 = df2.loc[df2["W"]==100]
ax1 = sns.lineplot(x="n", y="tiempo", data=dfW20);
ax1 = sns.lineplot(x="n", y="tiempo", data=dfW50);
ax1 = sns.lineplot(x="n", y="tiempo", data=dfW100);
ax1 = sns.lineplot(x="n", y="IS", data=df_cuadratico2);
ax1.legend(["W = 20", "W = 50", "W = 100", "0.9e-6.(2^(n))"]);
plt.ylabel("tiempo (ms)");
```

Vemos que si bien se trata de una cota superior, se encuentra por arriba de los distintos valores de $W$, acercándose más los valores mayores de $W$; esto pareciera que se condice con lo teorizado anteriormente. Sin embargo, para poder determinar si existe o no una correlación, grafiquemos la correlación y calculemos el coeficiente de Pearson para distintos valores de $W$ representativos que nos permita ver la correlación con la cota a medida que la capacidad máxima crece.

Por último, nos gustaría poder ver cómo afecta la poda de optimalidad a nuestro algoritmo. Para ello, pensamos un experimento complementario al anterior en el que los valores de beneficio $p_{i}$ varían entre 1 y 100, de forma que haya más beneficio disponible. Nuestra hipótesis acerca del resultado de esta experimentación es que al haber mayor rango de valores de beneficio, si se alcanza antes una suma de beneficio óptima, entonces se descartarán de ahí en adelante una mayor cantidad de casos para todos los subconjuntos que utilicen los valores más pequeños del rango de beneficios. Por lo tanto, esto resultará en una mayor cantidad de podas y, en consecuencia, una mejor performance y menores tiempos de ejecución. Presentamos a continuación un gráfico que compara ambas experimentaciones.

```python
# Graficamos el tiempo en función de n, con series variando m.
df4 = pd.read_csv("exp_todo_w_benefit_mayor_size_fijos_bt.csv")
ax1 = sns.lineplot(x="n", y="tiempo", data=df2);
ax1 = sns.lineplot(x="n", y="tiempo", data=df4);
ax1.legend(["experimentación beneficio [1,30]", "experimentación beneficio [1,100]"]);
plt.ylabel("tiempo (ms)");
```

El comportamiento que vemos en el gráfico no parece guardar relación con nuestra hipótesis planteada, ya que pareciera que en varios casos (a excepción del caso de $n=25$, en el que la relación se invierte) la experimentación nueva planteada termina teniendo un tiempo de ejecución peor que la experimentación anterior. Es decir, que un mayor rango de beneficios pareciera estar afectando negativamente a la performance. A raíz de estos nuevos resultados, podemos plantearnos una nueva hipótesis: a menor rango de beneficios otorgado, mejor será la performance (nótese que es inversa a la planteada inicialmente). Esto lo atribuimos a que cuanto mayor es el rango de beneficios (como vemos en el gráfico), más beneficio disponible habrá posiblemente en los elementos restantes que todavía no consideramos en nuestro subconjunto parcial. Por como funciona nuestra cota de optimalidad, siempre se compara si el beneficio que queda disponible más la suma actual de beneficios sobrepasa el máximo hasta el momento; si esto no se cumple, entonces se realiza la poda. Por lo tanto, al haber mayores valores de beneficios, hay posibilidad que uno de los últimos elementos a considerar sea uno de ellos, evitando que se realice la poda de optimalidad en un lugar considerablemente arriba del árbol de decisiones. Esto puede suceder, aparte, porque al principio del algoritmo se ordenan los elementos por tamaño decreciente (para mejorar la poda de factibilidad) pero no existe necesariamente una relación entre cómo crecen los tamaños con cómo crecen los beneficios.
Para poder determinar si esta hipótesis resulta correcta, probaremos esta vez disminuyendo aún más el rango de beneficios y chequearemos si los resultados se condicen con lo esperado. 

```python
df5 = pd.read_csv("exp_todo_w_benefit_menor_size_fijos_bt.csv")
ax1 = sns.lineplot(x="n", y="tiempo", data=df2.loc[df2["W"]==100]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df4.loc[df4["W"]==100]);
ax1 = sns.lineplot(x="n", y="tiempo", data=df5.loc[df5["W"]==100]);
ax1.legend(["experimentación beneficio [1,30]", "experimentación beneficio [1,100]", "experimentación beneficio [1,5]"]);
plt.ylabel("tiempo (ms)");
```
